# face landmark detection using dllib

>detect 68 landmarks



```
import dlib
import cv2
import numpy as np

from google.colab.patches import cv2_imshow 
```

```
img=cv2.imread("/content/drive/My Drive/face landmarks/cf.jpg")
h,w,ch=img.shape

img=cv2.resize(img,(int(w/2),int(h/2)))
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cv2_imshow(img)
```

***all 68 landmarks***

face68.png

```
#
detector=dlib.get_frontal_face_detector()
predictor=dlib.shape_predictor("/content/drive/My Drive/face landmarks/methods (fr,dllib)/dlib/shape_predictor_68_face_landmarks.dat")

#detect the faces
faces=detector(gray)

for face in faces:
  
  x1=face.left()
  y1=face.top()
  x2=face.right()
  y2=face.bottom()

  #landmark detection 
  landmarks=predictor(gray,face )
  
  #coordinates of point 0
  #x=landmarks.part(0).x
  #y=landmarks.part(0).y 

  for i in range(0,68):

    x=landmarks.part(i).x
    y=landmarks.part(i).y 

    cv2.circle(img,(x,y),2,(0,255,0),5)

cv2_imshow(img)
```
