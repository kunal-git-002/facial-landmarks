```
import cv2
import face_recognition as fr
import numpy as np
from google.colab.patches import cv2_imshow
```

```
def read_img(path):

  img=cv2.imread(path)
  h,w=img.shape[:2]
  new_width=800
  ratio=(new_width/w)

  new_height=int(h*ratio)

  img=cv2.resize(img,(new_width,new_height))

  return img

img=read_img("/content/drive/My Drive/cf.jpg")
```

1) finding face landmarks for single face

```
face_landmarks=fr.face_landmarks(img)

face_landmarks
```

```
for face_part in face_landmarks[0].keys():
  for points in face_landmarks[0][face_part]:

    cv2.circle(img,points,2,(0,255,0),-1)

cv2_imshow(img)
```

2) finding face landmark for multiple faces

```
img2=read_img("/content/drive/My Drive/rock2.jpg")
```

```
face_landmarks=fr.face_landmarks(img2)
```

```
for faces in face_landmarks:
  for face_part in faces.keys():
    for points in faces[face_part]:

      cv2.circle(img2,points,3,(0,255,0),-1)

cv2_imshow(img2)
```
